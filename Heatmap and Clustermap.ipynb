## Heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader as pdr ## https://pandas-datareader.readthedocs.io/en/latest/
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
## get some data
data_1 = pdr.get_data_google('AAPL')['Close'].pct_change()
data_2 = pdr.get_data_google('TSLA')['Close'].pct_change()
data_3 = pdr.get_data_google('BABA')['Close'].pct_change()
data_4 = pdr.get_data_google('FB')['Close'].pct_change()
data_5 = pdr.get_data_google('GOOG')['Close'].pct_change()

data = pd.DataFrame(data_1)
data.columns = ['AAPL']
data['TSLA'] = data_2
data['BABA'] = data_3
data['FB'] = data_4
data['GOOG'] = data_5
```

```python
data = data.dropna()
data.head()
```

```python
## heat map of correlation matrix
sns.heatmap(data.corr())
```

```python
## make this graph bigger
plt.figure(figsize = (10,10))
ax = sns.heatmap(data.corr())
```

```python
## rotate y label
ax = sns.heatmap(data.corr())
plt.yticks(rotation=0)
```

```python
## change color to other scheme
## all color scheme can be found in here: https://matplotlib.org/tutorials/colors/colormaps.html

ax = sns.heatmap(data.corr(),
                 cmap = 'YlGnBu')
plt.yticks(rotation=0)
```

```python
ax = sns.heatmap(data.corr(),
                 cmap = 'YlGnBu',
                 cbar_kws={"orientation": "horizontal"})
plt.yticks(rotation=0)

## set title and set font of title
ax.set_title("Correlation heatmap",
             fontsize=20,
             fontweight="bold")
plt.show()
```

```python
## add text, set annot = True
ax = sns.heatmap(data.corr(),
                 cmap = 'YlGnBu',
                 annot = True,
                 cbar_kws={"orientation": "horizontal"})
plt.yticks(rotation=0)

## set title and set font of title
ax.set_title("Correlation heatmap",
             fontsize=20,
             fontweight="bold")

plt.show()
```

```python
## clustermap
cg = sns.clustermap(data.corr(),
                    annot=True,
                    cmap = 'YlGnBu')
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(),rotation=0) ## rotate y label
plt.show()
```
